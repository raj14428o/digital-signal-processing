## Plot Unit Impulse Signal

**Definition:**

The unit step signal is defined as:

\
u(t) =
\begin{cases}
0, & t \ne 0 \\
1, & t = 0
\end{cases}

```python
import numpy as np 
import matplotlib.pyplot as plt
```

### Compute range of x using np.linspace()

```python
# Define x range
x = np.arange(-50 , 50, 10)
```

### Compute unit impulse signal

```python
y = np.where(x==0, 1, 0) 
```

### Plot unit impulse signal

```python

# Plot the discrete unit step signal
plt.stem(x, y)

# Add reference lines for axes
plt.axhline(y=0, color='k', linewidth=0.8)
plt.axvline(x=0, color='k', linewidth=0.8)

# Add labels and grid
plt.title("Unit impulse Signal")
plt.xlabel("x")
plt.ylabel("impulse(x)")
plt.grid(True)

# Set x-ticks -50 to 50 with step of 10
plt.xticks(np.arange(-50, 60, 10))

# Display the plot
plt.show()
```
