## Plot Unit Step Signal

**Definition:**

The unit step signal is defined as:

\
u(t) =
\begin{cases}
0, & t < 0 \\
1, & t \ge 0
\end{cases}

```python
import numpy as np 
import matplotlib.pyplot as plt
```

### Compute range of x using np.linspace()

```python
# Define x range
x = np.linspace(-50, 50, 40)
```

### Compute unit step values

```python
y = np.where(x >= 0, 1, 0)
```

### Plot unit step signal

```python

# Plot the discrete unit step signal
plt.stem(x, y)

# Add reference lines for axes
plt.axhline(y=0, color='k', linewidth=0.8)
plt.axvline(x=0, color='k', linewidth=0.8)

# Add labels and grid
plt.title("Unit Step Signal")
plt.xlabel("x")
plt.ylabel("u(x)")
plt.grid(True)

# Set x-ticks -50 to 50 with step of 10
plt.xticks(np.arange(-50, 60, 10))

# Display the plot
plt.show()
```
