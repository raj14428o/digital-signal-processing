## Dual Tone Multi Frequency (DTMF) Signal Generation

Dual Tone Multi-Frequency (DTMF) is generated by combining **two sinusoidal signals** of different frequencies —  
one from a **low-frequency group** and one from a **high-frequency group**.  
These two tones together represent a unique key on a telephone keypad.


```python
import numpy as np
import matplotlib.pyplot as plt
```

### Set Parameters

```python
# Sampling information
fs = 8000      # Sampling frequency in Hz
duration = 0.1 # Duration in seconds (20 ms)

# DTMF frequencies (example: digit '5' → 770 Hz + 1336 Hz)
f1 = 770     # Low frequency tone
f2 = 1336    # High frequency tone
```

### Define time vector and tones

```python
# Time vector
t = np.linspace(0, duration, int(fs * duration), endpoint=False)

# Generate the two tones
tone1 = np.sin(2 * np.pi * f1 * t)
tone2 = np.sin(2 * np.pi * f2 * t)

# Combine tones
combined_tone = tone1 + tone2
```

### Plot each tone

```python

# Plot the tones
plt.figure(figsize=(12, 8))

plt.subplot(3, 1, 1)
plt.plot(t, tone1, color='blue')
plt.title(f"Tone 1: sin(2π{f1}t)")
plt.xlabel("Time [s]")
plt.ylabel("Amplitude")

plt.subplot(3, 1, 2)
plt.plot(t, tone2 , color='green')
plt.title(f"Tone 2: sin(2π{f2}t)")
plt.xlabel("Time [s]")
plt.ylabel("Amplitude")

plt.subplot(3, 1, 3)
plt.plot(t, combined_tone , color='red')
plt.title(f"Combined DTMF Signal: sin(2π{f1}t) + sin(2π{f2}t)")
plt.xlabel("Time [s]")
plt.ylabel("Amplitude")

plt.tight_layout()
plt.show()
```
