## Problem Statement

The input signal \( x(n) = \{1, 2, 3, 1\} \) is applied to a discrete-time **Linear Time-Invariant (LTI)** system  
whose impulse response is given as \( h(n) = \{1, 2, 1, -1\} \).

The output of the system is obtained by the **discrete-time convolution**:

$$
y(n) = x(n) * h(n)
$$


```python
import numpy as np
import matplotlib.pyplot as plt
```

### Set input and impulse signal and compute output signal
### **Convolution**
The output of the system is given by:

$$
y(n) = \sum_{k=-\infty}^{\infty} x(k) \, h(n - k)
$$

```python
x = np.array([1,2,3,1])
h = np.array([1,2,1,-1])
y = np.convolve(x,h)
```

### Compute required auto correlation and cross correaltion
###  **Auto-correlation**
The auto-correlation of a discrete-time signal \( t(n) \) is defined as:

$$
R_{tt}(l) = \sum_{n=-\infty}^{\infty} t(n) \, t(n - l)
$$

### **Cross-correlation**
The cross-correlation between two signals \( x(n) \) and \( y(n) \) is given by:

$$
R_{xy}(l) = \sum_{n=-\infty}^{\infty} x(n) \, y(n - l)
$$

```python
# Autocorrelation of x(n)
r_xx = np.correlate(x, x, mode='full')

# Autocorrelation of h(n)
r_hh = np.correlate(h, h, mode='full')

# Autocorrelation of y(n)
r_yy = np.correlate(y, y, mode='full')

# Cross-correlation between x(n) and y(n)
r_xy = np.correlate(x, y, mode='full')
```

### Compute the time shift values

```python
# For a signal of length N, the time-shift index for auto-correlation
# ranges from -(N-1) to (N-1)
shift_x = np.arange(-(len(x) - 1), len(x))
shift_h = np.arange(-(len(h) - 1), len(h))
shift_y = np.arange(-(len(y) - 1), len(y))

# For cross-correlation between signals x (length N) and y (length M),
# the time-shift index ranges from -(N-1) to (M-1)
shift_xy = np.arange(-(len(x) - 1), len(y))
```

### Plot Autocorrelation of x(n)

```python

plt.figure(figsize=(12, 10))
plt.subplot(4, 1, 1)
plt.stem(shift_x, r_xx)
plt.title('Autocorrelation of Input Signal Rxx(l)')
plt.xlabel('Lag (l)')
plt.ylabel('Rxx(l)')
plt.grid(True)
```

### Plot Autocorrelation of h(n)

```python

plt.figure(figsize=(12, 10))
plt.subplot(4, 1, 2)
plt.stem(shift_h, r_hh)
plt.title('Autocorrelation of Impulse Response Rhh(l)')
plt.xlabel('Lag (l)')
plt.ylabel('Rhh(l)')
plt.grid(True)
```

### Plot Autocorrelation of y(n)

```python

plt.figure(figsize=(12,10))
plt.subplot(4, 1, 3)
plt.stem(shift_y, r_yy)
plt.title('Autocorrelation of Output Signal Ryy(l)')
plt.xlabel('Lag (l)')
plt.ylabel('Ryy(l)')
plt.grid(True)
```

### Plot Cross-correlation of x(n) and y(n)

```python

plt.figure(figsize=(12,10))
plt.subplot(4, 1, 4)
plt.stem(shift_xy, r_xy)
plt.title('Cross-correlation of x(n) and y(n) - Rxy(l)')
plt.xlabel('Lag (l)')
plt.ylabel('Rxy(l)')
plt.grid(True)
```
