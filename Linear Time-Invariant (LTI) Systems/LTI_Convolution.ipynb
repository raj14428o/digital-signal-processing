## The impulse response of LTI system is h(n) = {1,2,1,-1}. Determine the response of the system for input x(n) = {1,2,3,1} , y(n) = x(n).h(n)


```python
import numpy as np 
import matplotlib.pyplot as plt
```

### Set values of input and impulse signal

```python
x = np.array([1,2,1,-1])
h = np.array([1,2,3,1])
```

## Compute Output Signal

### Discrete-Time Convolution

For a discrete-time Linear Time-Invariant (LTI) system,  
the output signal **y(n)** is obtained by the **convolution** of the input **x(n)**  
with the system's impulse response **h(n)**:

\[
y(n) = x(n) * h(n)
\]

The convolution summation formula is given by:

$$
y(n) = \sum_{k=-\infty}^{\infty} x(k) \, h(n - k)
$$

where:

- \( x(k) \) → input signal  
- \( h(n - k) \) → time-shifted impulse response  
- \( y(n) \) → system output


```python
y = np.convolve(x,h)
```

### Plot input signal

```python
plt.subplot(3, 1, 1)
plt.stem(np.arange(len(x)), x)
plt.title("Input Signal x(n)")
plt.xlabel("n")
plt.ylabel("x(n)")
plt.grid(True)
plt.xticks(np.arange(0, len(x), 1));
```

### Plot impulse signal

```python

plt.subplot(3, 1, 2)
plt.stem(np.arange(len(h)), h)
plt.title("Impulse Response h(n)")
plt.xlabel("n")
plt.ylabel("h(n)")
plt.grid(True)
plt.xticks(np.arange(0, len(x), 1));
```

### Plot Output siganl

```python
plt.subplot(3, 1, 3)
plt.stem(np.arange(len(y)), y)
plt.title("Output Signal y(n) = x(n) * h(n)")
plt.xlabel("n")
plt.ylabel("y(n)")
plt.grid(True)
```
