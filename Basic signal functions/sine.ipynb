## Plot continuous and discrete sine wave

```python
import numpy as np
import matplotlib.pyplot as plt
```

### Generate continuous and discrete values of x on the plot

```python
# Continuous domain: 1000 points between -2π and 2π
x = np.linspace(-2 * np.pi, 2 * np.pi, 1000)

# Discrete domain: points spaced by 0.5 radians
x_disc = np.arange(-2 * np.pi, 2 * np.pi, 0.5)
```

### Compute sine wave for both continuous and discrete values

```python
y = np.sin(x)
y_disc = np.sin(x_disc)
```

### Plot continuous sine wave

```python
plt.title("Continuous Sine Wave")  # Set the title of the plot
plt.xlabel("x")                    # Label the x-axis
plt.ylabel("sin(x)")               # Label the y-axis
plt.grid(True)                     # Display grid lines on the plot  
plt.plot(x, y, label="sin(x)")     # plot continuous sine wave
plt.show()
```

### Plot Discrete sine wave

```python
plt.title("Continuous Sine Wave")  # Set the title of the plot
plt.xlabel("x")                    # Label the x-axis
plt.ylabel("sin(x)")               # Label the y-axis
plt.grid(True)                     # Display grid lines on the plot 
plt.stem(x_disc,y_disc)            # plot discrete sine wave
plt.show()
```
