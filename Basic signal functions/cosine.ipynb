## Plot continuous and discrete cosine wave

```python
import numpy as np
import matplotlib.pyplot as plt
```

### Generate discrete and continuous values of x on the plot

```python
#  Continuous domain: 1000 points between -2π and 2π
x = np.linspace(-2 * np.pi, 2 * np.pi, 1000)
# discrete domain : point space by 0.5 radians
x_disc = np.arange(-2*np.pi, 2*np.pi,0.5)
```

### Compute values of cosine for both continuous and discrete values

```python
y = np.cos(x)
y_disc = np.cos(x_disc)
```

### Plot continuous cosine wave

```python
plt.title("Continuous Cosine Wave")  # Set the title of the plot
plt.xlabel("x")                      # Label the x-axis
plt.ylabel("cos(x)")                 # Label the y-axis
plt.grid(True)                       # Display grid lines on the plot  
plt.plot(x, y, label="cos(x)")       # plot continuous cosine wave
plt.show()
```

### Plot discret cosine wave

```python
plt.title("Discrete Cosine Wave")         # Set the title of the plot
plt.xlabel("x")                           # Label the x-axis
plt.ylabel("cos(x)")                      # Label the y-axis
plt.grid(True)                            # Display grid lines on the plot  
plt.stem(x_disc, y_disc)                  # plot continuous sine wave
plt.show()
```
