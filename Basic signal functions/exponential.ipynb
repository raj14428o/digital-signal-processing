## Plot continuous and discrete exponential wave

```python
import numpy as np
import matplotlib.pyplot as plt
```

### Compute values of x for discrete and continuous plot

```python
#  Continuous domain: 1000 points between -2π and 2π
x = np.linspace(-2 * np.pi , 2 * np.pi , 1000)
# discrete domain : point space by 0.5 radians
x_disc = np.arange(-2 * np.pi , 2 * np.pi , 0.5)
```

### Compute values for exponential function

```python
exp = np.exp(x)
exp_decay = np.exp(-x)

exp_disc = np.exp(x_disc)
exp_disc_decay = np.exp(-x_disc)
```

### Plot continuous exponential wave

```python
plt.title("Continuous exponential growth")  # Set the title of the plot
plt.xlabel("x")                             # Label the x-axis
plt.ylabel("exp(x)")                        # Label the y-axis
plt.grid(True)                              # Display grid lines on the plot  
plt.plot(x, exp , label="exp(x)")           # plot continuous exponential growth
plt.show()
```

```python
plt.title("Continuous exponential growth")         # Set the title of the plot
plt.xlabel("x")                                    # Label the x-axis
plt.ylabel("exp(-x)")                              # Label the y-axis
plt.grid(True)                                     # Display grid lines on the plot  
plt.plot(x, exp_decay , label="exp(-x)")           # plot continuous exponential decay
plt.show()
```

### Plot discrete exponential wave

```python
plt.title("discrete exponential growth")    # Set the title of the plot
plt.xlabel("x")                             # Label the x-axis
plt.ylabel("exp(x)")                        # Label the y-axis
plt.grid(True)                              # Display grid lines on the plot  
plt.stem(x_disc , exp_disc)                 # plot discrete exponential growth
plt.show()
```

```python
plt.title("discrete exponential growth")          # Set the title of the plot
plt.xlabel("x")                                   # Label the x-axis
plt.ylabel("exp(-x)")                             # Label the y-axis
plt.grid(True)                                    # Display grid lines on the plot  
plt.stem(x_disc , exp_disc_decay)                 # plot discrete exponential decay
plt.show()
```
